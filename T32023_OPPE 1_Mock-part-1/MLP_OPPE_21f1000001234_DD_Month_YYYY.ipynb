```python
from sklearn.impute import SimpleImputer
```

```python
loc = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
est = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
carp = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
pro = SimpleImputer(missing_values='?', strategy='most_frequent')
```

```python
df['carpet_area'] = carp.fit_transform(pd.DataFrame(df['carpet_area']))
df['carpet_area']
```

```python
df['Property'] = pro.fit_transform(pd.DataFrame(df['Property']))
df['Property']
```

```python
df['Estimated Value'] = est.fit_transform(pd.DataFrame(df['Estimated Value']))
df['Estimated Value']
```

```python
df['Locality'] = loc.fit_transform(pd.DataFrame(df['Locality']))
df['Locality']
```
